```python
import pandas as pd

# Corrected URL
github_url = 'https://raw.githubusercontent.com/214G5A0502/Anil-Anudip_Project/main/Unemployment%20Data%20Analysis%20Project/data.csv'

# Read the CSV file from the corrected URL
df = pd.read_csv(github_url)

# Print the DataFrame to verify
print(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the data into a DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/214G5A0502/Anil-Anudip_Project/main/Unemployment%20Data%20Analysis%20Project/data.csv")

# Check the column names
print(df.columns)
```

```python
plt.figure(figsize=(10,6))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data[' Date'], region_data[' Estimated Unemployment Rate (%)'], label=region)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time by Region')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data[' Date'], region_data[' Estimated Employed'], label=region)
plt.xlabel('Date')
plt.ylabel('Employment Count')
plt.title('Employment Count Over Time by Region')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data[' Date'], region_data[' Estimated Labour Participation Rate (%)'], label=region)
plt.xlabel('Date')
plt.ylabel('Labour Participation Rate (%)')
plt.title('Labour Participation Rate Over Time by Region')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
freq_avg_unemployment = df.groupby(' Frequency')[' Estimated Unemployment Rate (%)'].mean()
freq_avg_unemployment.plot(kind='bar', color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Frequency Category')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df[' Estimated Unemployment Rate (%)'], df[' Estimated Employed'], alpha=0.5)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Employment Count')
plt.title('Relationship between Unemployment Rate and Employment Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df[' Estimated Labour Participation Rate (%)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Labour Participation Rate (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Labour Participation Rates Across All Regions')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['longitude'], df['latitude'], c=df[' Estimated Unemployment Rate (%)'], cmap='coolwarm', alpha=0.5)
plt.colorbar(label='Unemployment Rate (%)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Unemployment Rate by Geographical Location')
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_unemployment = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().nlargest(5)
top_unemployment.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Top 5 Regions with Highest Unemployment Rates')
plt.show()
```

```python
plt.figure(figsize=(10,6))
for freq in df[' Frequency'].unique():
    freq_data = df[df[' Frequency'] == freq]
    plt.plot(freq_data[' Date'], freq_data[' Estimated Unemployment Rate (%)'], label=freq)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time by Frequency Category')
plt.legend()
plt.show()
```

```python
region = 'YourRegion'
region_data = df[df['Region'] == region]
plt.figure(figsize=(10,6))
plt.plot(region_data[' Date'], region_data[' Estimated Unemployment Rate (%)'], marker='o')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title(f'Unemployment Rate Over Time in {region}')
plt.show()
```

```python
frequency_category = 'Monthly'
frequency_data = df[df[' Frequency'] == frequency_category]
plt.figure(figsize=(10,6))
plt.plot(frequency_data[' Date'], frequency_data[' Estimated Labour Participation Rate (%)'], marker='o')
plt.xlabel('Date')
plt.ylabel('Labour Participation Rate (%)')
plt.title(f'Labour Participation Rate Over Time for {frequency_category} Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    plt.plot(region_data[' Date'], region_data[' Estimated Unemployment Rate (%)'], label=region)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time Across Regions')
plt.legend()
plt.show()
```
